```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the problem
v0 = 1  # Initial velocity (arbitrary units)
m = 1    # Mass (arbitrary units)
c = 0.1  # Linear drag coefficient (arbitrary units)
k = 0.1  # Quadratic drag coefficient (arbitrary units)

# Time array
t = np.linspace(0, 10, 1000)

# Velocity as a function of time under linear drag
v_linear = v0 / (1 + c * v0 * t / m)

# Displacement as a function of time under linear drag (integral of velocity)
x_linear = np.cumsum(v_linear) * (t[1] - t[0])

# Velocity as a function of time under quadratic drag
v_quadratic = v0 * np.exp(-k * t)

# Displacement as a function of time under quadratic drag (integral of velocity)
x_quadratic = np.cumsum(v_quadratic) * (t[1] - t[0])

# Plotting the displacement under linear and quadratic drag
plt.figure(figsize=(10, 6))
plt.plot(t, x_linear, label='Linear Drag', color='blue')
plt.plot(t, x_quadratic, label='Quadratic Drag', color='red')

# Adding labels, title, and saving the plot
plt.xlabel('Time')
plt.ylabel('Displacement')
plt.title('Displacement vs. Time under Linear and Quadratic Drag')
plt.legend()
plt.grid(True)
plt.savefig("extra_credit_1.png", dpi=500)
plt.show()
```
